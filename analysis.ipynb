## Get json data from the API

```python
import requests, pandas as pd, time

startTime = time.time()

# get sale data
response = requests.get('https://borrowcupapi.herokuapp.com/api/sale')
saleData = response.json()

print('Took', round(time.time() - startTime, 2), 'seconds to get Sale data.')
startTime = time.time()

# get return data
response = requests.get('https://borrowcupapi.herokuapp.com/api/return')
returnData = response.json()

print('Took', round(time.time() - startTime, 2), 'seconds to get Return data.')
```

## Convert sale data to dataframe

```python
saleDF = pd.DataFrame.from_dict(saleData, orient='columns')
#saleDF.head()
```

## Drop irrelevant columns and rename some others

```python
saleDF = saleDF.drop(['scanned_at', 'scanned_at_melbourne_date_time'], axis=1)
saleDF = saleDF.rename(columns={"scanned_at_melbourne_date": "sale_date", "scanned_at_melbourne_time": "sale_time", "id": "sale_id"})
saleDF.head()
```

## Convert return data to dataframe

```python
from pandas.io.json import json_normalize
returnDF = pd.DataFrame.from_dict(json_normalize(returnData), orient='columns')
#returnDF.head()
```

## Drop irrelevant columns and rename some other

```python
returnDF = returnDF.drop(['scanned_at', 'bin_id', 'cafe_id', 'dishwasher_id'], axis=1)
returnDF = returnDF.rename(columns={"scanned_at_melbourne_date_time.date": "return_date", "scanned_at_melbourne_date_time.time": "return_time", "id": "return_id"})
returnDF.head()
```

## Merge sale and return dataframes (full-outer join)

```python
mergedDF = saleDF.merge(returnDF, how='outer', on='return_id')
mergedDF.head(20)
```

## Get sale date-time in one column and change column data-type to _dateTime64_

```python
mergedDF['date_time_concat'] = mergedDF['sale_date'] + ' ' + mergedDF['sale_time']
mergedDF['sale_date_time'] = pd.to_datetime(mergedDF['date_time_concat'], dayfirst=True)
mergedDF = mergedDF.drop(['sale_date', 'sale_time', 'date_time_concat'], axis=1)
mergedDF.head()
```

## Get return date-time in one column and change column data-type to _dateTime64_

```python
mergedDF['date_time_concat'] = mergedDF['return_date'] + ' ' + mergedDF['return_time']
mergedDF['return_date_time'] = pd.to_datetime(mergedDF['date_time_concat'], dayfirst=True)
mergedDF = mergedDF.drop(['return_date', 'return_time', 'date_time_concat'], axis=1)
mergedDF.head()
```

## Change data-type of some columns to Integer from Float denoting NaNs by zero and add a flag column - _modified_ - with zeroes for all rows

### Modified flag status:

| Flag  | Meaning |
| :---: | :---: |
| `0`  | Sale and return info are both actual and nothing is inferred  |
| `1`  | Sale date-time is inferred and is not actual  |
| `2`  | Return date-time is inferred and is not actual  |

```python
mergedDF.fillna({'cafe_id': 0, 'cup_id_x': 0, 'sale_id': 0, 'return_id': 0, 'cup_id_y': 0}, inplace=True, downcast='infer')
mergedDF['modified'] = 0
mergedDF.tail()
```

```python
def compare(row):
    if not pd.isna(row['sale_date_time']) and not pd.isna(row['return_date_time']):
        if row['sale_date_time'] < row['return_date_time']:
            return row['sale_date_time']
        else:
            return row['return_date_time']
    elif pd.isna(row['sale_date_time']):
        return row['return_date_time']
    else:
        return row['sale_date_time']

mergedDF['timestamp'] = mergedDF.apply(lambda row: compare(row), axis = 1)
mergedDF = mergedDF.sort_values(by='timestamp')
mergedDF
```

## Infer scan date-times of sales/returns where they are unknown

#### Process:
1. For each row, identify whether return or sale info is missing
2. If return info is missing:
    - Get cup id and sale scan date-time
    - For that cup id, try to find the next sale recorded
    - If next sale is found:
        - Get next sale date-time
        - Find time difference between the cup's current sale and the next sale found.
        - If the next sale happened within 6 hours:
            - It is probably a case of repeated scans by cafe staff and the current sale record should be discarded. Thus, add that row index to the `drop_rows` list.
        - If the next sale happened after more than 6 hours:
            - We will infer that a return between the two sales must have happened and set a return date-time of a time 2 hours before the next sale date-time to the record at hand.
            - Set the modified flag to `2` to denote that the return date-time is inferred and not actual.

3. If sale info is missing:
    - Get cup id and return scan date-time
    - For that cup id, try to find the previous return recorded
    - If previous return is found:
        - Get previous return date-time
        - Find time difference between the cup's previous return found and the current return.
        - If the previous return happened within 18 hours:
            - It is improbable that the cup would have sold and returned so quickly and the current record should be discarded. Thus, add that row index to the `drop_rows` list.
        - If the previous return happened after more than 18 hours:
            - We will infer that a sale between the two return must have happened and set a sale date-time of 3 P.M. for the date next to the previous return date to the record at hand.
            - Set the modified flag to `1` to denote that the sale date-time is inferred and not actual.

```python
drop_rows = []
for index, row in mergedDF.iterrows():
    if row['return_id'] == 0:
        cup_id = row['cup_id_x']
        sale_date_time = row['sale_date_time']
        filteredDF = mergedDF[(mergedDF['cup_id_x'] == cup_id) & (mergedDF['sale_date_time'] > sale_date_time)].sort_values(by='sale_date_time')
        if len(filteredDF.index) > 0:
            next_sale_date_time = filteredDF.iloc[0, 5]
            time_diff = next_sale_date_time - sale_date_time

            # check time difference between the 2 scans
            if time_diff <= pd.Timedelta(hours=6):

                # potential case of a repeated scan, thus we will remove that record
                drop_rows.append(index)
            else:
                mergedDF.loc[index, 'return_date_time'] = next_sale_date_time - pd.Timedelta(hours=2)
                mergedDF.loc[index, 'modified'] = 2
    elif row['sale_id'] == 0:
        cup_id = row['cup_id_y']
        return_date_time = row['return_date_time']
        filteredDF = mergedDF[(mergedDF['cup_id_y'] == cup_id) & (mergedDF['return_date_time'] < return_date_time)].sort_values(by='return_date_time', ascending = False)
        if len(filteredDF.index) > 0:
            previous_return_date_time = filteredDF.iloc[0, 6]
            time_diff = return_date_time - previous_return_date_time
            
            # check time difference between the 2 scans
            if time_diff <= pd.Timedelta(hours=18):

                # potential case of a repeated scan, thus we will remove that record
                drop_rows.append(index)
            else:
                mergedDF.loc[index, 'sale_date_time'] = (previous_return_date_time + pd.Timedelta(days=1)).replace(hour=15, minute=0, second=0)
                mergedDF.loc[index, 'modified'] = 1
drop_rows
```
